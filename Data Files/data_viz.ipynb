# Cleaning for datawrapper

```python
# I found that there are some zip codes in my dataset that DW says it can't match with / doesn't identify as NYC zip code. I want to check those again here.

invalid_zips = [
    "10279", "10162", "10803", "11021", "11791", "11747", "11580",
    "11735", "11023", "11010", "10704", "11706", "11787", "10710",
    "14068", "12524", "11553", "11575", "10977", "11520", "14845",
    "10523", "11721", "11801", "12180", "11560", "11779", "11788",
    "10970", "11542", "11767", "14223", "11516", "11590", "10591",
    "11552", "12203", "11793", "11733", "11950", "11003", "11570",
    "11727", "10598", "11581", "10573", "11803", "10701"
]
```

```python
# find me every zip code in my file that is in the invalid zips
df[df["zip"].astype(str).str.zfill(5).isin(invalid_zips)] #convert to string, add padding if needed, and check it is in invalid zips
```

```python
# dropping these values
df = df[
    ~df["zip"].astype(str).str.zfill(5).isin(invalid_zips)
].copy()
```

```python
#confirming values got dropped
df[df["zip"].astype(str).str.zfill(5).isin(invalid_zips)]
```

## Chloropleth map 1: data with zip and complaint counts

```python
df.shape
```

```python
df
```

```python
# Creating zip level dataset

df_zip = df.groupby("zip").agg(
    complaint_count=("unique_key", "count"),
    pop=("pop", "first"),
    median_hh_income=("median_hh_income", "first"),
    income_category=("income_category", "first")
).reset_index()
```

```python
df_zip
```

```python
# creating complaints per 100K column 

df_zip["complaints_per_100k"] = (
    df_zip["complaint_count"] / df_zip["pop"] * 100000
)
```

```python
# converting complaint column to whole numbers 
df_zip["complaints_per_100k"] = df_zip["complaints_per_100k"].round().astype(int)
```

```python
df_zip
```

```python
(df_zip["complaints_per_100k"] < 40000).value_counts(normalize=True) * 100
```

```python
# saving dataset 
df_zip.to_csv("complaintsper100k_by_zip.csv", index=False)
```

## Chloropleth map 2: with only top complaints

```python
# calculating 90th percentile for complaints per 100k to put into datawrapper
df_zip["complaints_per_100k"].quantile(0.9)
```

## Bar graph 1: median complaints per 100k by income category

```python
df_zip["complaints_per_100k"] = df_zip["complaints_per_100k"].round().astype(int)
```

```python
df_complaints_by_income = (
    df_zip
    .groupby("income_category", as_index=False)["complaints_per_100k"]
    .median()
)
```

```python
df_complaints_by_income
```

```python
# saving dataset 
df_complaints_by_income.to_csv("bar_graph_complaints_by_income.csv", index=False)
```
