**The variables we have:**

- when the complaint was closed (Month/Date/Year Hour:Minutes:Seconds AM / PM)

***What***
- what the complaint was (this include problem, problem detail, additional details)

***Where***
- which borough the complaint came from
- which city the complaint came from (which is the same as borough, largely)
- exact latitude and longitude of complaint
- which zip code the complaint came from (this is for merging with income data)

***Response***
- Status (values include -> )
- If resolved, when did it get updated in the dataset
- What happened in the resolution

***Misc***
- unique key of each complaint (each row = one complaint)
- Agency to which it was submitted to

## Categorizing Income Data

```python
# checking data type of income column 
df_complaints["median_hh_income"].dtype
#string
```

```python
# on converting this column to int, it did not convert, which means there is some values which might be like 250K +. 
df_complaints["median_hh_income"].value_counts().head(20)
```

```python
# finding odd values in income
df_complaints["median_hh_income"].unique()
```

```python
# checking odd values in income 

bad_values = df_complaints.loc[
    pd.to_numeric(df_complaints["median_hh_income"], errors="coerce").isna(), # in rows where you're unable to convert values into numeric values
    "median_hh_income"
]

bad_values.value_counts()
```

```python
# converting 250,000+ to 250000, removing commas, and marking - as missing values 

df_complaints["median_hh_income"] = (
    df_complaints["median_hh_income"]
    .astype(str) # convert to string so str. operations can work
    .str.replace(",", "", regex=False) #removing ,
    .str.replace("+", "", regex=False) #removing +
    .replace("-", pd.NA) #pd.NA means treat this value as missing
)

df_complaints["median_hh_income"] = pd.to_numeric(
    df_complaints["median_hh_income"],
    errors="coerce"
)
```

```python
# checking range of income column now (it is already a float)
df_complaints["median_hh_income"].describe()


# Among complaints that matched to income data:
# 25% are from ZIP codes with median household income below about $59,700.
# 50% are from ZIP codes with median household income below about $77,600.
# 75% are from ZIP codes with median household income below about $97,500.
```

For deciding income categories, I should use ZIP-level income, not complaint-level income. Otherwise high-complaint ZIPs influence the cutoffs.

```python
# creating dataset with only unique zip and income levels (same as our income data from ACS) to decide income categorisations:
zip_income = (
    df_complaints[["zip", "median_hh_income"]]
    .drop_duplicates()
    .dropna()
)
```

```python
zip_income
```

```python
# Income quartiles are based on ZIP-level median household income,
# using one row per ZIP code with valid ACS income data. 
zip_income["median_hh_income"].quantile([0, 0.25, 0.5, 0.75, 1])

# Q1: $74,185
# Q2 / median: $95,345
# Q3: $127,956
```

```python
# Lowest income:       $24,086 – $74,185
# Lower-middle income: $74,185 – $95,345
# Upper-middle income: $95,345 – $127,956
# Highest income:      $127,956 – $250,000
```

```python
bins = [0, 74184.75, 95344.50, 127955.75, 250000]

labels = [
    "Lowest income",
    "Lower-middle income",
    "Upper-middle income",
    "Highest income"
]

df_complaints["income_category"] = pd.cut(
    df_complaints["median_hh_income"],
    bins=bins,
    labels=labels,
    include_lowest=True
)
```

```python
df_complaints["income_category"].value_counts(normalize=True, dropna=False)*100
```

Most complaints in my matched dataset come from ZIPs in the lowest income quartile. But that could be because:
- lower-income ZIPs have more people,
- lower-income ZIPs have more 311-reportable problems,
- lower-income residents complain more

```python
#making broader buckets for complaints

def group_complaints(complaint):
    
    if complaint in [
        "Noise - Residential",
        "Noise - Street/Sidewalk",
        "Noise - Commercial",
        "Noise",
        "Noise - Vehicle"
    ]:
        return "noise"

    elif complaint in [
        "HEAT/HOT WATER",
        "Water System",
        "PLUMBING",
        "WATER LEAK",
        "Sewer"
    ]:
        return "drainage, water and plumbing"

    elif complaint in [
        "UNSANITARY CONDITION",
        "Dirty Condition",
        "Illegal Dumping",
        "Missed Collection"
    ]:
        return "dirty or unsanitary conditions"

    elif complaint in [
        "Illegal Parking",
        "Blocked Driveway",
        "Abandoned Vehicle",
        "Derelict Vehicles"
    ]:
        return "parking and vehicles"

    else:
        return complaint.lower()
```

- noise:
Combines residential noise, street/sidewalk noise, commercial noise, vehicle noise, and general noise complaints. These all refer to unwanted sound disturbances, even though the source differs.

- housing, water and plumbing:
Combines HEAT/HOT WATER, Water System, PLUMBING, WATER LEAK, and Sewer. These categories all relate to basic housing infrastructure and water-related service failures. The grouping is broad, but useful because hot water, water supply, leaks, plumbing, and sewer issues overlap in lived experience.

- dirty or unsanitary conditions:
Combines UNSANITARY CONDITION, Dirty Condition, Illegal Dumping, and Missed Collection. These categories all relate to trash, pests, mold, garbage, dumping, waste removal, or unsanitary physical conditions. They are combined because they describe cleanliness/sanitation problems in buildings or public space.

- parking and vehicles:
Combines Illegal Parking, Blocked Driveway, Abandoned Vehicle, and Derelict Vehicles. These all relate to vehicle storage, obstruction, or improper vehicle use in public/private space.

```python
#applying buckets to dataset
df_complaints_clean["complaint_group"] = df_complaints_clean["complaint"].apply(group_complaints)
```
